```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import slackweb
import io

def america_index():
    url = 'https://www.traders.co.jp/market_fo/'
    res = requests.get(url)
    soup = bs(res.text, "html.parser")
    info = soup.findAll(class_='zone')[1]

    dow = "ＮＹダウ"
    sp500 = "Ｓ＆Ｐ５００"
    nasdaq = "NASDAQ"

    dow_v = info.findAll(class_='text-right')[2].text.replace( '\n' , '' )
    sp500_v = info.findAll(class_='text-right')[8].text.replace( '\n' , '' )
    nasdaq_v = info.findAll(class_='text-right')[14].text.replace( '\n' , '' )

    data_set = {dow: dow_v, sp500: sp500_v, nasdaq : nasdaq_v}
    return data_set

def sp500_sector():
    url = 'https://www.traders.co.jp/market_fo/sector_ranking'
    res = requests.get(url)
    soup = bs(res.text, "html.parser")
    info = soup.findAll(class_='zone')[0]
    info_s = info.find_all(class_='data_table table')

    sector_sp = []
    for x in range(1, 52, 5):
        for info in info_s:
            info = info.find_all("td")[x].text.replace(" ","")
            info = info.replace('\n' , '' )
            sector_sp.append(info)

    sp_vola = []
    for info_v in info_s:
        info = info_v.find_all("span")
        for vola in info:
            if vola.text[-1] == '%':
                vola = vola.text
                sp_vola.append(vola)

    sector_info_list = dict(zip(sector_sp, sp_vola))
    return sector_info_list

def main():
    america = america_index()
    vola = sp500_sector()
    america.update(vola)

    df = pd.DataFrame(america.values(), index=america.keys())
    name = ["volatility"]
    df.columns = name

    #Slackのwebhookurlを変更する
    slack_url = "https://hooks.slack.com/xxxx"

    tsv = io.StringIO()
    df.to_csv(tsv, sep='\t')

    slack = slackweb.Slack(url=slack_url)
    slack.notify(text=tsv.getvalue())

if __name__ == '__main__':
    main()
```

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import slackweb
import io

def market_data():

    url = 'https://www.traders.co.jp/market_jp/'
    res = requests.get(url)

    soup = bs(res.text, "html.parser")
    info = soup.findAll(class_='zone')[1]

    nikkei = "日経平均"
    topix = "TOPIX"
    mothers = "マザーズ"

    nikkei_volatility = info.findAll(class_='text-right')[2].text.replace( '\n' , '' )
    topix_volatility = info.findAll(class_='text-right')[8].text.replace( '\n' , '' )
    mothers_volatility = info.findAll(class_='text-right')[14].text.replace( '\n' , '' )

    data_set = {nikkei: nikkei_volatility, topix: topix_volatility, mothers: mothers_volatility}
    return data_set

def sector_volatility():
    url = 'https://www.traders.co.jp/market_jp/sector_ranking/day'
    res = requests.get(url)

    soup = bs(res.text, "html.parser")
    info = soup.findAll(class_='zone')

    sector = []
    for i in info:
        Industry = i.find_all(class_='text-nowrap')
        for i in Industry:
            Industry = i.text
            sector.append(Industry)

    sector = sector[1:]
    pro_sector = []
    for change in sector:
        sector = change.replace("\n\n                                                    ","")
        sector = sector.replace("（東証）\n                                                \n\n\n\n","")
        pro_sector.append(sector)

    sector_vola = []
    for test in info:
        Industry_v = test.find_all(class_='text-right')
        for i in Industry_v:
            Industry_v = i.find_all("span")
            if len(Industry_v) != 0 and  Industry_v[0].text[-1] == '%':
                for i in Industry_v:
                    i = i.text
                    sector_vola.append(i)

    dict_from_list = dict(zip(pro_sector, sector_vola))
    return dict_from_list

def main():
    market = market_data()
    vola = sector_volatility()
    market.update(vola)

    df = pd.DataFrame(market.values(), index=market.keys())
    name = ["volatility"]
    df.columns = name

    #Slackのwebhookurlを変更する
    slack_url = "https://hooks.slack.com/xxxx"

    tsv = io.StringIO()
    df.to_csv(tsv, sep='\t')

    slack = slackweb.Slack(url=slack_url)
    slack.notify(text=tsv.getvalue())

if __name__ == '__main__':
    main()
```
